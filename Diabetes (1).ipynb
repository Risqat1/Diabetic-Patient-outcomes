```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python

data['DiabetesPedigreeFunction'] = sc.fit_transform(data['DiabetesPedigreeFunction'].values.reshape(-1,1))
```

```python
from sklearn.preprocessing import OneHotEncoder
ONE = OneHotEncoder()
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = data.drop('Outcome', axis = 1) # X here represents our features(independent variable)
y = data['Outcome'] # y here represents our labels(Dependent Variable)
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size = 0.10, random_state = 30)
```

```python
X
```

```python
X_train
```

```python
y_train
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion ='gini',random_state = 20)
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred3 = dtc.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred3)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred3))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred3))
```

```python
data.to_csv('diabetes', index = False)
```
